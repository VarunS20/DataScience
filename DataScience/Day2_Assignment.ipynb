***Questions 1:***
Create an empty list. Accept 10 numbers from the user and append to it the list if it is an even number.*italicized text*

```
inList = list()
for i in range(11):
  inp = input('Enter a number')
  inList.append(int(inp))

li = list()
for i in inList:
  i = int(i)
  if i%2==0:
    li.append(i)
print(li)
```

**Questions 2:**Create a notebook on LIST COMPREHENSION. This exercise is to put you in a Self learning mode


List comprehensions provide us with a simple way to create a list based on some iterable. During the creation, elements from the iterable can be conditionally included in the new list and transformed as needed.

The components of a list comprehension are:

    

1.   Output Expression (Optional)
2.   Iterable
3.   Iterator variable which represents the members of the iterable





   
     

```
# printing the even number in the range provided as a input
n = int(input())
li = [ i for i in range(n) if i%2==0 ]
print(li)
```

```
numbers = [1, 2, 3, 4, 5]
squares = [number**2 for number in numbers]
print(squares)
```

```
#List in List
board = [[i*i for i in range(4)] for j in range(4)]
print(board)
```

**Question 3:-**You have seen in the videos how powerful dictionary data structure is.
In this assignment, given a number n, you have to write a program that generates a dictionary d which
contains (i, i*i), where i is from 1 to n (both included).
Then you have to just print this dictionary d.

```
dic = {}
inp = int(input())
for i in range(inp):
  if i==0:
    continue
  dic[i]=i*i
print(dic)
```

**Question4:**There is a robot which wants to go the charging point to charge itself.
The robot moves in a 2-D plane from the original point (0,0). The robot can
move toward UP, DOWN, LEFT and RIGHT with given steps.

```
x1=0
y1=0
ndir = int(input())
for i in range(ndir):
  move = input()
  direction = move.split()[0].upper()
  steps = int(move.split()[1])
  if direction == 'UP':
    y1 = y1+steps
  elif direction == 'DOWN':
    y1 = y1-steps
  elif direction == 'LEFT':
    x1=(x1*-1)-steps
  elif direction == 'RIGHT':
    x1=x1+steps
  else:
    print("Invalid Input")
    exit()
displacement = (x1**2+y1**2)**0.5
displacement = round(displacement)
print(displacement)
```
