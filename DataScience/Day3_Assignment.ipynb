**Questions 4:**
How to convert a 1D array into a 2D array? Demonstrate with the help of a code snippet
Hint: np.newaxis, np.expand_dims

Converting of an 1D array to 2D array can be done using two ways
1. using numpy.newaxis()
2. using numpy.expand_dims()


1. Using numpy.newaxis:- The np.newaxis is just an alias for the Python constant None, which means that wherever you use np.newaxis you could also use None:

**How to use:**The np.newaxis is generally used with slicing. It indicates that you want to add an additional dimension to the array. The position of the np.newaxis represents where I want to add dimensions.

```
import numpy as np
arr = np.array([1,2,3,4,5,6,7])
print(arr)
print(arr.shape)
arr1=arr[:, np.newaxis]
print(arr1)
print(arr1.shape)
arr2 = arr[np.newaxis,:]
print(arr2)
print(arr2.shape)
```

2. Using numpy.expand_dims():- The second method is to use numpy.expand_dims() function that has an intuitive axis kwarg. This function takes two parameters. The first is the array of which you want to increase the dimension of and the second is index/indexes of array on which you want to create a new axis.









```
import numpy as np
x = np.zeros((3, 4)) 
print(x)
print(x.shape)
y = np.expand_dims(x, axis=1) 
print(y) 
print(y.shape)
```

How to get unique items and counts of unique items?

Ans: Python’s numpy library provides a numpy.unique() function to find the unique elements and it’s corresponding frequency in a numpy array.
and returns Sorted unique elements of an array with their corresponding frequency counts NumPy array.

```
import numpy as np
arr = np.array([1,1,1,1,3,3,2,2,4,5,67,89,900,00,100,100,4])
unique,count =np.unique(arr,return_counts=True)
print('unique elements are ',unique)
print('Counts of the above elements are in the same order',count) 
```
